# Equazioni di primo grado

Questo programma risolve l'equazione di primo grado $ax+b=0$ con $a,b\in\mathbb{R}$ parametri reali, ed $x\in\mathbb{R}$ incognita.

## Definizione delle funzioni

```python
def input_parametri_equazione_grado_1():
    print("Questo programma risolve l'equazione: a x + b = 0")
    a = float(input("Inserisci il valore del parametro a: "))
    b = float(input("Inserisci il valore del parametro b: "))
    return (a,b)
    
def soluzione_equazione_grado_1(a,b):
    if a == 0 and b == 0:
        print("Indeterminata")
        return
    elif a == 0 and b != 0:
        print("Impossibile")
        return
    else:
        return -b/a

def risolvi_equazione_grado_1():
    a,b = input_parametri_equazione_grado_1()
    x = soluzione_equazione_grado_1(a,b)
    if x != None:
        print(f"x = {x}")
```

## Uso delle funzioni

```python
risolvi_equazione_grado_1()
```

```python
x = soluzione_equazione_grado_1(3.14159, 1.4142)
if x != None:
    print(f"x = {x}")
```

```python
x = soluzione_equazione_grado_1(0,0)
if x != None:
    print(f"x = {x}")
```

```python
x = soluzione_equazione_grado_1(0,1)
if x != None:
    print(f"x = {x}")
```
