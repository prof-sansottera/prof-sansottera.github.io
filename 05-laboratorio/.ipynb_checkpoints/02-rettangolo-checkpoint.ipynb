# Rettangolo

## Definizione delle funzioni

```{warning} Attenzione
Quando scrivi un programma, è bene *non* mischiare le _definizioni_ delle funzioni con il resto del codice dove le funzioni _vengono effettivamente utilizzate_. 
```

In questa sezione definiamo delle funzioni, che poi useremo in seguito.  Nota bene che quando definisci una funzione, stai semplicemente insegnando al computer ad associare al nome di una funzione, una serie di operazioni da eseguire (un _algoritmo_),  *non* gli stai dicendo di _usare_ la funzione.

````python
def area_rettangolo(a, b):
    area = a*b
    return area

def perimetro_rettangolo(a, b):
    perimetro = 2*(a+b)
    return perimetro

def area_perimetro_rettangolo(a, b):
    area = area_rettangolo(a, b)
    perimetro = perimetro_rettangolo(a, b)
    return (area, perimetro) # restituisce una tupla
````

## Uso delle funzioni

Dopo aver definito le funzioni, possiamo ora usarle per calcolare quel che ci interessa.  In questo caso calcoliamo area e perimetro di un rettangolo fissando il valore della base a 2 e dell'altezza a 5.  Come vedi, i nomi delle variabili non devono essere uguali a quelli dei parametri nella definizione della funzione.

```{warning}Python ti da abbastanza corda per impiccarti da solo...
**NON** usare mai lo stesso nome per una variabile e una funzione!
```

````python
base = 2
altezza = 5

area = area_rettangolo(base, altezza)
print("L'area del rettangolo di lati", base, "e", altezza, "è", area)

perimetro = perimetro_rettangolo(base, altezza)
print("Il perimetro del rettangolo di lati", base, "e", altezza, "è", perimetro)

area2, perimetro2 = area_perimetro_rettangolo(base, altezza)
print(f"base = {base}, altezza = {altezza}, area = {area2}, perimetro = {perimetro2}")
````
