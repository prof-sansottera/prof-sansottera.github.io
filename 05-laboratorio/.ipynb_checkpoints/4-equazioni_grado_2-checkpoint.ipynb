# Equazioni di secondo grado

Questo programma risolve l'equazione di secondo grado $ax^2+bx+c=0$ con $a,b,c\in\mathbb{R}$ parametri reali, ed $x\in\mathbb{R}$ incognita.

## Import librerie

```python
import math
```

## Definizione delle funzioni

```python
def input_parametri_equazione_grado_2():
    print("Questo programma risolve l'equazione: a x^2 + b x + c = 0")
    a = float(input("Inserisci il valore del parametro a: "))
    b = float(input("Inserisci il valore del parametro b: "))
    c = float(input("Inserisci il valore del parametro c: "))
    return (a,b,c)
    
def soluzione_equazione_grado_1(a,b):
    if a == 0 and b == 0:
        print("Indeterminata")
        return
    elif a == 0 and b != 0:
        print("Impossibile")
        return
    else:
        return -b/a

def soluzione_equazione_grado_2(a,b,c):
    if a == 0:
        x = soluzione_equazione_grado_1(b,c)
        if x == None:
            return
        else:
            return [x] # Restituisco una lista con una sola soluzione, per poter poi usare len(x)!
    else:
        delta = b**2-4*a*c
        if delta < 0:
            print("Impossibile")
            return
        else:
            x1 = (-b+math.sqrt(delta))/(2*a)
            x2 = (-b-math.sqrt(delta))/(2*a)
            return [x1, x2]

def risolvi_equazione_grado_2():
    a,b,c = input_parametri_equazione_grado_2()
    x = soluzione_equazione_grado_2(a,b,c)
    if x != None:
        if len(x) == 1:
            print(f"x = {x[0]}")
        else:
            print(f"x1 = {x[0]}, x2 = {x[1]}")
```

## Uso delle funzioni

```python
risolvi_equazione_grado_2()
```
