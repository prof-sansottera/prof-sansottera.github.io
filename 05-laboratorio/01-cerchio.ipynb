# Area e perimetro del cerchio

## Definizione delle funzioni

```{warning} Attenzione
Quando scrivi un programma, è bene *non* mischiare le _definizioni_ delle funzioni con il resto del codice dove le funzioni _vengono effettivamente utilizzate_. 
```

In questa sezione definiamo delle funzioni, che poi useremo in seguito.  Nota bene che quando definisci una funzione, stai semplicemente insegnando al computer ad associare al nome di una funzione, una serie di operazioni da eseguire (un _algoritmo_),  *non* gli stai dicendo di _usare_ la funzione.

````python
def area_cerchio(raggio):
    pi = 3.14159 # valore approssimato di Pi greco
    area = pi*raggio**2
    return area

def perimetro_cerchio(raggio):
    pi = 3.14159
    perimetro = 2*pi*raggio
    return perimetro

def area_perimetro_cerchio(raggio):
    area = area_cerchio(raggio)
    perimetro = perimetro_cerchio(raggio)
    return (area, perimetro) # restituisce una tupla
````

## Uso delle funzioni

Dopo aver definito le funzioni, possiamo ora usarle per calcolare quel che ci interessa.  In questo caso calcoliamo area e circonferenza del cerchio fissando il valore del raggio a 2.  Come vedi, il nome della variabile non deve essere uguale a quello del parametro nella definizione della funzione.

```{warning}Python ti da abbastanza corda per impiccarti da solo...
**NON** usare mai lo stesso nome per una variabile e una funzione!
```

````python
r = 3.0

area = area_cerchio(r)
print("L'area del cerchio di raggio", r, "è", area)

circonferenza = perimetro_cerchio(r)
print("La circonferenza del cerchio di raggio", r, "è", circonferenza)

area2, circonferenza2 = area_perimetro_cerchio(r)
print(f"R = {r}, Area = {area2}, Circonferenza = {circonferenza2}")
````
